### Where SEM Sits Between the Paradigms

- SEM is the **Model First** paradigm pushed to its logical conclusion: the analyst draws the full causal diagram **a priori** — arrows, latent variables, everything.
- Because the model never peeks at the data, classical inference is valid on the **full sample** — no split required.
- But one data-driven choice — a feature, a square, a freed path — silently moves us to **Data First**, and Lecture 2's rules take over.
- Along the way we will see that correlation, regression, path analysis, factor analysis, and simultaneous equations are **all special cases of one framework**.

<div style="border-left: 5px solid #003D79; background-color: #F5F7FA; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #003D79; font-family: 'Georgia', 'Palatino Linotype', serif; margin-top: 0;">The General SEM: Measurement + Structural Model</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7; margin-bottom: 0;">A structural equation model has two layers. The <strong>structural (latent) model</strong> relates latent variables to each other; the <strong>measurement model</strong> says how latent variables produce the observed indicators.</p>
</div>

**Structural (latent) model** — relations among latent variables:

$$\boldsymbol{\eta} \;=\; \mathbf{B}\,\boldsymbol{\eta} \;+\; \boldsymbol{\Gamma}\,\boldsymbol{\xi} \;+\; \boldsymbol{\zeta}$$

**Measurement model** — how latent variables produce observed indicators:

$$\mathbf{y} \;=\; \boldsymbol{\Lambda}_y\,\boldsymbol{\eta} + \boldsymbol{\varepsilon},
\qquad
\mathbf{x} \;=\; \boldsymbol{\Lambda}_x\,\boldsymbol{\xi} + \boldsymbol{\delta}$$

where:

- $\boldsymbol{\eta}$ — endogenous latent variables (explained inside the system);
- $\boldsymbol{\xi}$ — exogenous latent variables (taken as given);
- $\boldsymbol{\Lambda}_y, \boldsymbol{\Lambda}_x$ — factor loadings;
- $\mathbf{B}$ — paths among endogenous latents; it may contain **feedback loops** (simultaneous equations!);
- $\boldsymbol{\zeta}, \boldsymbol{\varepsilon}, \boldsymbol{\delta}$ — structural and measurement errors.

<div style="background-color: #162F57; padding: 18px 22px; margin: 20px 0; border-radius: 8px;">
  <p style="margin: 0; color: #FFF; font-weight: 700; font-family: 'Georgia', 'Palatino Linotype', serif;">The Central Idea: Covariance-Structure Modeling</p>
  <p style="margin: 10px 0 0 0; color: #E8EEF5; line-height: 1.6;">The parameters $\theta$ imply a model covariance matrix $\Sigma(\theta)$. Estimation chooses $\hat{\theta}$ so that $\Sigma(\hat{\theta})$ is as close as possible to the sample covariance matrix $\mathbf{S}$. Everything observable about a linear SEM lives in the covariances — "fitting" means reconciling the theoretical and empirical covariance matrices.</p>
</div>

### Estimation: Maximum Likelihood

Under multivariate normality, maximum likelihood minimizes the discrepancy

$$F_{ML}(\theta) \;=\; \log|\Sigma(\theta)| + \mathrm{tr}\!\left(\mathbf{S}\,\Sigma(\theta)^{-1}\right) - \log|\mathbf{S}| - p$$

- $F_{ML} = 0$ exactly when $\Sigma(\hat{\theta}) = \mathbf{S}$: the model reproduces the sample covariances perfectly.
- Standard ML asymptotics give standard errors from the information matrix: $\hat{\theta} \approx \mathcal{N}\big(\theta,\ \mathcal{I}^{-1}/n\big)$.

### Testing the Specified Structure

- Overall fit test: $T = (n-1)\,F_{ML}(\hat{\theta}) \sim \chi^2_{df}$ under $H_0$: *the specified structure is correct*.
- $df = \frac{p(p+1)}{2} - (\#\text{free parameters})$ — the number of **over-identifying restrictions**.
- Practical fit indices: CFI, TLI, RMSEA, SRMR, AIC/BIC.

<div style="background-color: #FBEAED; border-left: 4px solid #C4122F; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #C4122F; font-weight: 700;">Notice What Is Being Tested</p>
  <p style="margin: 10px 0 0 0; color: #333;">A <strong>pre-specified</strong> $H_0$, on the <strong>whole sample</strong>. The sampling distribution of $T$ is valid <em>because the model did not peek at the data</em>. Keep this sentence in mind — it is the hinge of the whole lecture.</p>
</div>

### The <code>semopy</code> Syntax — Three Operators Cover Everything

`semopy` uses lavaan-style model syntax inside a Python string:

| Operator | Example | Meaning |
|:---------|:--------|:--------|
| `~`  | `y ~ x1 + x2` | regression (structural path) |
| `=~` | `F =~ y1 + y2 + y3` | measurement: latent `F` loads on indicators |
| `~~` | `x1 ~~ x2` | (co)variance between variables / errors |

Directed arrows (`~`, `=~`) are **causal claims you assert**; `~~` is a mere **association you allow**.

### Example 1 — A First Fit: Specify, Fit, Report

The whole inferential workflow is three lines: specify the model as a string, fit on the **full sample**, and report estimates and fit statistics.

```python
adv = load_data('Advertising.csv', index_col=0)

desc  = "sales ~ TV + radio + newspaper"
model = semopy.Model(desc)
model.fit(adv)                        # ML estimation on the FULL sample

print(model.inspect())                # estimates, SEs, z, p-values
print()
print(semopy.calc_stats(model).T)     # chi2, df, CFI, RMSEA, AIC, BIC ...
```

### One Framework, Five Familiar Faces

Each classical model is obtained by **fixing some SEM matrices to zero or identity**:

| Classical model | SEM restriction | Latents? | Dataset today |
|:----------------|:----------------|:--------:|:--------------|
| Correlation | saturated covariances, no paths | no | `Advertising.csv` |
| Regression | one endogenous $y$ | no | `Advertising.csv` |
| Path analysis | chained regressions | no | `Credit.csv` |
| Factor analysis (CFA) | measurement model only | yes | `bfi.csv` |
| Simultaneous equations | non-recursive $\mathbf{B}$ | no | `Carseats.csv` |
| Full SEM | everything at once | yes | `acsi_data.csv` |

### Example 2 — Special Case 1: Correlation

A **saturated** covariance model: 3 free parameters (two variances, one covariance), 3 sample moments, so $df = 0$ — the model reproduces $\mathbf{S}$ exactly and cannot be rejected. The **standardized** covariance estimate is Pearson's $r$, and its $p$-value is the classical test of $H_0: \rho = 0$ — computed on **all** $n = 200$ markets.

```python
desc = """
TV ~~ sales     # free covariance
TV ~~ TV        # variances
sales ~~ sales
"""
m_corr = semopy.Model(desc)
m_corr.fit(adv)
print(m_corr.inspect(std_est=True))

r_np = np.corrcoef(adv['TV'], adv['sales'])[0, 1]
print(f"\nnumpy corrcoef check: r = {r_np:.4f}")
# The std. estimate of 'TV ~~ sales' equals Pearson's r ~ 0.78
```

### Example 3 — Special Case 2: Regression

No latents, one endogenous variable: SEM–ML **reproduces OLS**. We verify against `statsmodels`. This is the famous ISLR result: `newspaper` is not significant *given* TV and radio — classic inferential output on **all** the data.

```python
m_reg = semopy.Model("sales ~ TV + radio + newspaper")
m_reg.fit(adv)
print("--- semopy (SEM-ML) ---")
print(m_reg.inspect().head(4))

ols = smf.ols('sales ~ TV + radio + newspaper', data=adv).fit()
print("\n--- statsmodels OLS ---")
print(ols.params.round(4))
print("\nSame estimates: regression is an SEM with one endogenous variable.")
```

### Example 4 — Special Case 3: Path Analysis (Mediation)

Marginally, $r(\text{Income}, \text{Balance}) = 0.46$ but $r(\text{Limit}, \text{Balance}) = 0.86$: does the credit **limit** mediate the income effect? Theory dictates Income $\to$ Limit $\to$ Balance — the data cannot choose the arrows' direction, **you** assert it beforehand.

- Indirect effect $= a \cdot b$; total effect $= c' + a\cdot b$.
- Watch the sign of the direct effect $c'$: it comes out **negative** although the marginal correlation is positive — a classic **suppression** pattern (at a fixed credit limit, higher income means balances are paid down faster).

```python
credit = load_data('Credit.csv')

desc = """
Limit   ~ Income
Balance ~ Limit + Income
"""
m_path = semopy.Model(desc)
m_path.fit(credit)
est = m_path.inspect()
print(est)

a  = est.loc[(est.lval == 'Limit')   & (est.rval == 'Income'), 'Estimate'].iloc[0]
b  = est.loc[(est.lval == 'Balance') & (est.rval == 'Limit'),  'Estimate'].iloc[0]
c_ = est.loc[(est.lval == 'Balance') & (est.rval == 'Income'), 'Estimate'].iloc[0]
print(f"\nindirect effect a*b = {a*b:8.3f}")
print(f"direct   effect c'  = {c_:8.3f}   (negative: suppression!)")
print(f"total    effect     = {c_ + a*b:8.3f}")
```

### Example 5 — Special Case 4: Confirmatory Factor Analysis

Big Five inventory, ~2,800 respondents, items scored 1–6. The **two-factor theory** (Extraversion, Neuroticism) is written down before the data: each latent loads only on its own five items, and the two factors may correlate. Items E1 and E2 are reverse-worded — expect **negative** loadings.

The $\chi^2$ (df = 34) tests this two-factor structure. Contrast with *exploratory* factor analysis, which searches for the number of factors — an **adaptive choice** we return to in Part 5.

```python
bfi = load_data('bfi.csv')
items = ['E1','E2','E3','E4','E5','N1','N2','N3','N4','N5']
bfi_en = bfi[items].dropna()

desc = """
Extraversion =~ E1 + E2 + E3 + E4 + E5
Neuroticism  =~ N1 + N2 + N3 + N4 + N5
Extraversion ~~ Neuroticism
"""
m_cfa = semopy.Model(desc)
m_cfa.fit(bfi_en)
print(semopy.calc_stats(m_cfa).T)
print()
print(m_cfa.inspect(std_est=True).head(12))   # loadings: E1, E2 negative
```

### Example 6 — Special Case 5: Simultaneous Equations

Feedback: price drives sales (demand), and sales feed back into pricing — a **non-recursive** $\mathbf{B}$:

$$\begin{aligned}
\text{Sales} &= \beta_1\,\text{Price} + \gamma_1\,\text{Income} + \gamma_2\,\text{Advertising} + \zeta_1 \\
\text{Price} &= \beta_2\,\text{Sales} + \gamma_3\,\text{CompPrice} + \zeta_2
\end{aligned}$$

Identification comes from **exclusion restrictions** (instruments) fixed by economic theory a priori: Income and Advertising shift demand but not pricing; CompPrice shifts pricing but not demand. The econometric supply–demand tradition is just an SEM with observed variables.

```python
cars = load_data('Carseats.csv')

desc = """
Sales ~ Price + Income + Advertising  # demand
Price ~ Sales + CompPrice             # pricing
"""
m_sim = semopy.Model(desc)
m_sim.fit(cars)
print(m_sim.inspect())

naive = smf.ols('Sales ~ Price', data=cars).fit()
print(f"\nnaive OLS price effect : {naive.params['Price']:.4f}")
print("SEM (system) price effect: see 'Sales ~ Price' above —")
print("the difference is simultaneity bias.")
```

### Example 7 — Full SEM: The ACSI Model

The *American Customer Satisfaction Index* is itself a famous SEM: regression + factor analysis + path analysis in a **single likelihood**. We load the survey, keep the indicator columns, clean the missing codes, and fit measurement + structural model at once. The global $\chi^2$ asks: *could these covariances arise from this structure?*

```python
acsi = (load_data('acsi_data.csv', na_values='#NULL!')
          [["OVERALLQ","CUSTOMQ","WRONGQ","SATIS","CONFIRM","IDEAL",
            "REPUR","HIGHPTOL"]]
          .apply(pd.to_numeric, errors="coerce")
          .dropna())
print(f"{len(acsi)} complete responses")

desc = """
# --- measurement model ---
Quality      =~ OVERALLQ + CUSTOMQ + WRONGQ
Satisfaction =~ SATIS + CONFIRM + IDEAL
Loyalty      =~ REPUR + HIGHPTOL
# --- structural model ---
Satisfaction ~ Quality
Loyalty      ~ Satisfaction
"""
m_acsi = semopy.Model(desc)
m_acsi.fit(acsi)
print(semopy.calc_stats(m_acsi).T)      # global chi2 test of the WHOLE theory
print()
print(m_acsi.inspect().head(12))
```

<div style="background-color: #162F57; padding: 18px 22px; margin: 20px 0; border-radius: 8px;">
  <p style="margin: 0; color: #FFF; font-weight: 700; font-family: 'Georgia', 'Palatino Linotype', serif;">The Inferential Contract</p>
  <p style="margin: 10px 0 0 0; color: #E8EEF5; line-height: 1.6;">The assumption behind every standard error, $p$-value, and $\chi^2$ test: the model structure $\mathcal{M}$ (arrows, variables, functional form) was fixed <strong>before</strong> looking at the data. The data are used <em>once</em>, only to estimate $\theta$ and test $\mathcal{M}$. Under this contract there is <strong>no model search</strong> — hence no risk of "overfitting the search". Sampling theory already quantifies estimation uncertainty.</p>
</div>

### Why Splitting Would Actually *Hurt*

- Splitting would only **discard information**: with a 50/50 split every standard error grows by roughly $\sqrt{2}$ — and nothing is gained, because there was no search to protect against.
- The honest "validation" of a fixed model is **replication on a new sample** (Cobb–Douglas was re-fit on new decades!) …
- … and the **over-identification $\chi^2$ test** itself: the model's restrictions can already fail on the full sample.

### Example 8 — Watching the Standard Errors Inflate

Fit the same pre-specified mediation model on the full `Credit` sample and on a random half. Same estimates (in expectation), fatter standard errors — pure loss.

```python
desc = """
Limit   ~ Income
Balance ~ Limit + Income
"""

m_full = semopy.Model(desc)
m_full.fit(credit)
full = m_full.inspect().query("op == '~'")[['lval','rval','Estimate','Std. Err']]

half_idx = rng.choice(len(credit), size=len(credit)//2, replace=False)
m_half = semopy.Model(desc)
m_half.fit(credit.iloc[half_idx])
half = m_half.inspect().query("op == '~'")[['lval','rval','Estimate','Std. Err']]

cmp = full.merge(half, on=['lval','rval'], suffixes=('_full','_half'))
cmp['SE ratio'] = cmp['Std. Err_half'] / cmp['Std. Err_full']
print(cmp.round(4))
print(f"\nSE inflation ~ sqrt(2) = {np.sqrt(2):.3f} — information thrown away for nothing.")
```

### Why Does Fixing the Model Make Full-Sample Inference Valid?

A $p$-value is a promise about *hypothetical replications*:

$$p = \Pr\big(\,T \ge t_{obs}\ \big|\ H_0,\ \mathcal{M}\ \text{fixed}\,\big)$$

- If $\mathcal{M}$ is fixed, the distribution of $T$ over replications is the advertised $\chi^2$/normal — the promise holds.
- If $\mathcal{M}$ was *selected* because it looked good on these data, the same selection rule would pick a **different** model in each replication — the advertised null distribution no longer applies.

### The Garden of Forking Paths

- Even **one** data-driven decision — drop a variable, add a square, free a covariance — silently changes the reference distribution of every test statistic downstream.
- The analyst may never have run more than one model, yet the *potential* forks (what they **would** have done had the data looked different) already invalidate the nominal $p$-values.

### The Three-Way Split (Lecture 2's Machinery Returns)

| Portion | Share | Role |
|:--------|:-----:|:-----|
| **Train** | 60% | estimate parameters of *every* candidate (degree $1,\dots,K$; every feature subset; …) |
| **Validation** | 20% | compare candidates on unseen data; choose the winner (or use $k$-fold CV) |
| **Test** | 20% | report the winner's performance on data unseen by **both** fitting and selection — used **once** |

### Example 9 — Polynomial Degree on `Wage.csv`: the *Wrong* Way

Selecting the degree by in-sample fit on **all** the data: the in-sample error falls monotonically with $K$, so the loop crowns wiggly, overfit curves. Worse, reporting the $p$-values of the *winning* degree as if it were pre-specified is invalid — they were computed on the same data that crowned the winner.

```python
wage = load_data('Wage.csv')
for K in range(1, 8):
    wage[f'age{K}'] = wage['age'] ** K

def sem_mse(model, df, target='wage'):
    """Out-of-sample MSE of a fitted semopy model."""
    yhat = model.predict(df.drop(columns=[target]))[target]
    return float(np.mean((df[target] - yhat) ** 2))

# WRONG: select degree by in-sample fit on ALL the data --------------
insample = {}
for K in range(1, 8):
    cols = ' + '.join(f'age{k}' for k in range(1, K + 1))
    m = semopy.Model(f'wage ~ {cols}')
    m.fit(wage)                              # full sample
    insample[K] = sem_mse(m, wage)           # scored on the SAME data

print('in-sample MSE by degree (monotone decreasing -> always picks big K):')
for K, v in insample.items():
    print(f'  K = {K}:  {v:8.2f}')
```

### Example 10 — The Right Way: Split *Before* Any Model Is Fit

A 60/20/20 split with `random_state=310`, done **before** the search. Each candidate degree is fit on **train** and judged on **validation** — never on the data it was fit to. The test set enters exactly once, at the very end: its MSE is the only unbiased estimate of real-world error.

```python
from sklearn.model_selection import train_test_split

train, hold = train_test_split(wage, test_size=0.4, random_state=310)
val, test   = train_test_split(hold, test_size=0.5, random_state=310)
print(f'train {len(train)} / validation {len(val)} / test {len(test)}')

scores_tr, scores_val = {}, {}
for K in range(1, 8):
    cols = ' + '.join(f'age{k}' for k in range(1, K + 1))
    m = semopy.Model(f'wage ~ {cols}')
    m.fit(train)                             # TRAIN only
    scores_tr[K]  = sem_mse(m, train)
    scores_val[K] = sem_mse(m, val)          # VALIDATION only

K_star = min(scores_val, key=scores_val.get)  # hyperparameter chosen

final = semopy.Model('wage ~ ' + ' + '.join(f'age{k}' for k in range(1, K_star + 1)))
final.fit(train)
print(f'\nchosen degree K* = {K_star}')
print(f'honest test MSE  = {sem_mse(final, test):.2f}   (TEST, used ONCE)')
```

### The Validation Curve

Train error keeps falling; validation error is U-shaped — the bias–variance trade-off of Lecture 2. $K^\star$ minimizes *validation* error: a decision made by data, i.e. a hyperparameter.

```python
Ks = sorted(scores_val)
fig, ax = plt.subplots()
ax.plot(Ks, [scores_tr[K]  for K in Ks], 'o-', color=BC_BLUE, label='train MSE')
ax.plot(Ks, [scores_val[K] for K in Ks], 's-', color=BC_RED,  label='validation MSE')
ax.axvline(K_star, color=BC_SILVER, ls='--', label=f'$K^*$ = {K_star}')
ax.set_xlabel('polynomial degree $K$ (flexibility)')
ax.set_ylabel('MSE')
ax.set_title('Underfit \u2190 \u2192 Overfit: the Validation Curve')
ax.legend()
fig.text(0.99, 0.01, 'Source: Wage.csv — CS 310, Bellevue College',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
```

### Example 11 — Feature Selection on `Hitters.csv`

Five candidate predictors of `Salary` ⟹ $2^5 - 1 = 31$ possible models — an aggressive search. With enough candidates, *some* subset fits noise well by luck; only validation exposes it. The $p$-values printed for the winning subset on the training data are **not** valid inferential statements: selection has occurred.

```python
from itertools import combinations

hit = load_data('Hitters.csv').dropna(subset=['Salary'])
features = ['AtBat', 'Hits', 'Walks', 'Years', 'PutOuts']

train_h, hold_h = train_test_split(hit, test_size=0.4, random_state=310)
val_h, test_h   = train_test_split(hold_h, test_size=0.5, random_state=310)

results = {}
for r in range(1, len(features) + 1):
    for subset in combinations(features, r):
        m = semopy.Model('Salary ~ ' + ' + '.join(subset))
        m.fit(train_h[list(subset) + ['Salary']])          # TRAIN
        results[subset] = sem_mse(m, val_h[list(subset) + ['Salary']],
                                  target='Salary')          # VALIDATION

best_subset = min(results, key=results.get)
print(f'{len(results)} candidate models searched')
print(f'best subset on validation: {best_subset}')

final_h = semopy.Model('Salary ~ ' + ' + '.join(best_subset))
final_h.fit(train_h[list(best_subset) + ['Salary']])
print(f'honest test MSE: {sem_mse(final_h, test_h[list(best_subset) + ["Salary"]], target="Salary"):.0f}')
```

### Example 12 — K-Fold Cross-Validation Instead of a Single Split

A single validation split is noisy; $k$-fold CV averages the selection signal over $k$ rotations of the held-out fold (Lecture 2). The test set still stays untouched until the very end.

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=310)
trval = pd.concat([train, val])                # test still untouched

cv_scores = {}
for K in range(1, 8):
    cols = ' + '.join(f'age{k}' for k in range(1, K + 1))
    fold_mse = []
    for tr_idx, va_idx in kf.split(trval):
        m = semopy.Model(f'wage ~ {cols}')
        m.fit(trval.iloc[tr_idx])
        fold_mse.append(sem_mse(m, trval.iloc[va_idx]))
    cv_scores[K] = float(np.mean(fold_mse))
    print(f'K = {K}:  5-fold CV MSE = {cv_scores[K]:8.2f}')

K_cv = min(cv_scores, key=cv_scores.get)
final_cv = semopy.Model('wage ~ ' + ' + '.join(f'age{k}' for k in range(1, K_cv + 1)))
final_cv.fit(trval)
print(f'\nCV-chosen degree K* = {K_cv}')
print(f'honest test MSE     = {sem_mse(final_cv, test):.2f}   (used once)')
```

### The Same Trap Inside SEM: Modification Indices

- SEM software can suggest which fixed parameter, if freed, would most improve $\chi^2$ (**modification indices**).
- Freeing parameters one by one until the model "fits" is **stepwise feature selection in disguise**.
- After such respecification, the final $\chi^2$ and $p$-values no longer have their nominal distributions on the same data.

### One Table to Remember

| | **Model First (SEM)** | **Data First (adaptive)** |
|:--|:--|:--|
| Model structure | fixed a priori by theory | searched (degree, features, MIs) |
| Data usage | full sample, once | train / validation / test |
| Validity of SEs, $p$'s | classical theory applies | invalid after selection |
| Overfitting control | parsimony + $\chi^2$ over-id. test | holdout / cross-validation |
| "Validation" means | replication, over-id. test | out-of-sample error |

### Three Takeaways

1. Correlation, regression, path models, CFA, and simultaneous equations are **all SEMs** — restrictions of one covariance-structure framework.
2. A pre-specified model needs **no** train/validation split: sampling theory handles uncertainty; splitting wastes data.
3. Any data-driven choice — features, polynomial degree, transformations, freed paths — is a **hyperparameter**: select it on validation data (or by cross-validation), report honest error on an untouched test set — or use a confirmatory sample for inference.
