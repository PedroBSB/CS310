  <p style="font-size: 1.05em; color: #555; line-height: 1.7; margin-bottom: 0;">Late 1920s: economist Paul Douglas and mathematician Charles Cobb seek a simple law relating <strong>output</strong> $Y$ to <strong>capital</strong> $K$ and <strong>labor</strong> $L$. E.g. $Y$ = shoes produced; $K$ = machinery and equipment funding; $L$ = workers employed. Nearly a century later, still central to micro, macro, and growth theory.</p>

### Structural Assumptions

1. Output is generated from capital $K$ and labor $L$.
2. Output responds **smoothly and monotonically** to changes in inputs — no jumps.
3. The **elasticity** of output with respect to each input is **constant**: a 1% increase in $K$ always raises $Y$ by the same percentage.
4. Baseline case: **constant returns to scale** — double both inputs, double the output.

### Assumptions Become Equations

The constant-elasticity assumption says: each input's *marginal* productivity is proportional to its *average* productivity. Formally, this is a system of **partial differential equations**:

$$\frac{\partial Y}{\partial K}=\alpha\frac{Y}{K},
\qquad
\frac{\partial Y}{\partial L}=\beta\frac{Y}{L},
\qquad \alpha,\beta>0.$$

<div style="background-color: #162F57; padding: 18px 22px; margin: 20px 0; border-radius: 8px;">
  <p style="margin: 0; color: #FFF; font-weight: 700; font-family: 'Georgia', 'Palatino Linotype', serif;">The Inferential Moment</p>
  <p style="margin: 10px 0 0 0; color: #E8EEF5; line-height: 1.6;"><strong>Before any data are seen</strong>, the theory has already pinned down the mathematics. Everything that follows — the functional form, the meaning of every parameter — is a consequence of the assumptions, not of a search over model classes.</p>
</div>

### Solving the System, Steps 1–3

**Step 1 — Separate variables in the $K$-equation.** Start with $\dfrac{\partial Y}{\partial K}=\alpha\dfrac{Y}{K}$ and divide both sides by $Y$:

$$\frac{1}{Y}\frac{\partial Y}{\partial K}=\frac{\alpha}{K}.$$

Now $Y$ appears only on the left, $K$ only on the right — ready to integrate.

**Step 2 — Integrate with respect to $K$**, holding $L$ fixed:

$$\int \frac{1}{Y}\frac{\partial Y}{\partial K}\,dK=\int \frac{\alpha}{K}\,dK
\quad\Longrightarrow\quad
\log Y=\alpha\log K+f(L).$$

**Key subtlety:** the "constant" of integration can depend on $L$, since $L$ was held fixed — hence an arbitrary **function** $f(L)$, not a number.

**Step 3 — Exponentiate.**

$$Y(K,L)=K^{\alpha}\exp\{f(L)\}.$$

Progress: the dependence on $K$ is fully determined; only the unknown factor $\exp\{f(L)\}$ remains. To find $f$, we bring in the **second** equation of the system.

### Solving the System, Steps 4–5

**Step 4 — Impose the $L$-equation.** Differentiate the current form with respect to $L$:

$$\frac{\partial Y}{\partial L}=K^{\alpha}\exp\{f(L)\}\,f'(L).$$

The theory requires $\dfrac{\partial Y}{\partial L}=\beta\dfrac{Y}{L}$, so

$$K^{\alpha}\exp\{f(L)\}\,f'(L)=\beta\,\frac{K^{\alpha}\exp\{f(L)\}}{L}
\quad\Longrightarrow\quad
f'(L)=\frac{\beta}{L}.$$

**Step 5 — Solve for $f$ and assemble.** Integrating $f'(L)=\beta/L$ gives $f(L)=\beta\log L+\log A$, where $\log A$ is a genuine constant. Substituting back:

$$\boxed{\;Y(K,L)=A\,K^{\alpha}L^{\beta}\;}$$

<div style="background-color: #162F57; padding: 18px 22px; margin: 20px 0; border-radius: 8px;">
  <p style="margin: 0; color: #FFF; font-weight: 700; font-family: 'Georgia', 'Palatino Linotype', serif;">The Cobb–Douglas Production Function</p>
  <p style="margin: 10px 0 0 0; color: #E8EEF5; line-height: 1.6;">$Y=A\,K^{\alpha}L^{\beta}$, with $A>0$ the <em>total factor productivity</em>. The form is <strong>not ad hoc</strong>: it is the unique consequence of the assumptions.</p>
</div>

### Step 6 — Check the Returns to Scale

Scale both inputs by $t>0$:

$$Y(tK,tL)=A(tK)^{\alpha}(tL)^{\beta}=t^{\alpha+\beta}\,Y(K,L).$$

- $\alpha+\beta=1$: **constant** returns to scale.
- $\alpha+\beta<1$: **decreasing** returns.
- $\alpha+\beta>1$: **increasing** returns.

The solution *verifies* the original assumption — a built-in sanity check.

### Example 2 — Let the Computer Redo the Calculus

`sympy` verifies symbolically that $Y = A K^{\alpha} L^{\beta}$ satisfies **both** partial differential equations, and rederives the returns-to-scale identity. The theory checks out — by machine.

```python
K, L, A, alpha, beta, t = sp.symbols('K L A alpha beta t', positive=True)
Y = A * K**alpha * L**beta

pde_K = sp.simplify(sp.diff(Y, K) - alpha * Y / K)
pde_L = sp.simplify(sp.diff(Y, L) - beta * Y / L)
print('dY/dK - alpha*Y/K  simplifies to:', pde_K)
print('dY/dL - beta *Y/L  simplifies to:', pde_L)

scale = sp.simplify(Y.subs({K: t * K, L: t * L}) / Y)
print('Y(tK, tL) / Y(K, L)          =', scale)

# And sympy can even solve the K-equation from scratch (L held fixed):
Yfun = sp.Function('Y')
ode = sp.Eq(Yfun(K).diff(K), alpha * Yfun(K) / K)
print('General solution of the K-equation:', sp.dsolve(ode).rhs,
      '   <- C1 plays the role of exp{f(L)}')
```

### Example 3 — The Production Surface and Its Isoquants

The function for $\alpha = 0.25$, $\beta = 0.75$, $A = 1.01$ (spoiler: the values Cobb and Douglas found). Each **isoquant** collects all $(K, L)$ combinations producing the same output — and doubling both inputs doubles output, as constant returns promise.

```python
A_, al_, be_ = 1.01, 0.25, 0.75
Kg, Lg = np.meshgrid(np.linspace(20, 400, 200), np.linspace(20, 400, 200))
Yg = A_ * Kg**al_ * Lg**be_

fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))
cs = axes[0].contour(Kg, Lg, Yg, levels=[50, 100, 150, 200, 250, 300],
                     colors=BC_BLUE)
axes[0].clabel(cs, fmt='Y = %d', fontsize=9)
axes[0].plot([100, 200], [100, 200], 'o--', color=BC_RED, linewidth=2, markersize=8)
axes[0].annotate('double both inputs\n(100,100) -> (200,200)', xy=(200, 200),
                 xytext=(230, 130), color=BC_RED, fontsize=10,
                 arrowprops=dict(arrowstyle='->', color=BC_RED))
axes[0].set_xlabel('capital $K$')
axes[0].set_ylabel('labor $L$')
axes[0].set_title('Isoquants of $Y = 1.01\,K^{0.25}L^{0.75}$')

ax3d = fig.add_subplot(1, 2, 2, projection='3d')
axes[1].remove()
ax3d.plot_surface(Kg, Lg, Yg, cmap='Blues', alpha=0.9)
ax3d.set_xlabel('$K$')
ax3d.set_ylabel('$L$')
ax3d.set_zlabel('$Y$')
ax3d.set_title('The Production Surface', color=BC_BLUE, fontweight='bold')

fig.text(0.99, -0.02, 'Source: CS 310 Lecture 3.', ha='right', fontsize=9, color=BC_SILVER)
plt.tight_layout()
plt.show()

Y1 = A_ * 100**al_ * 100**be_
Y2 = A_ * 200**al_ * 200**be_
print(f'Y(100, 100) = {Y1:.2f}   Y(200, 200) = {Y2:.2f}   ratio = {Y2 / Y1:.4f}')
print('Doubling both inputs exactly doubles output: constant returns (a + b = 1).')
```

$$\log Y=\log A+\alpha\log K+\beta\log L+\varepsilon.$$

- $\varepsilon$: productivity shocks, measurement error, deviations from the idealized theory.
- Data **estimate** $(A,\alpha,\beta)$ and **test** hypotheses, e.g. $H_0:\alpha+\beta=1$.
- **The data do not select the model** — they quantify and validate a theory.

  <p style="margin: 10px 0 0 0; color: #E8EEF5; line-height: 1.6;">In Lecture 2 we split the data because the <em>data themselves</em> chose the model, and we needed untouched observations to keep the choice honest. Here the model was fixed by theory <strong>before</strong> the first observation — there is no selection to protect against, so <strong>all 24 observations</strong> go into the estimates. Honesty is enforced by the standard errors and the model's assumptions, not by a held-out set.</p>
</div>

### Example 5 — ERM on All the Data

Empirical risk minimization under squared loss on the log-linear form — ordinary least squares, exactly as in Lectures 1–2, but on **every observation**. Compare the estimates with the values Cobb and Douglas made famous: $\alpha \approx 1/4$, $\beta \approx 3/4$.

```python
cd['logY'] = np.log(cd.Y)
cd['logK'] = np.log(cd.K)
cd['logL'] = np.log(cd.L)

model = smf.ols('logY ~ logK + logL', data=cd).fit()
print(model.summary())

alpha_hat = model.params['logK']
beta_hat = model.params['logL']
A_hat = np.exp(model.params['Intercept'])
print(f'\nEstimated production function:  '
      f'Y = {A_hat:.3f} * K^{alpha_hat:.3f} * L^{beta_hat:.3f}')
print(f'alpha_hat + beta_hat = {alpha_hat + beta_hat:.4f}  '
      '(constant returns would be exactly 1)')
```

### Example 6 — Inference: Confidence Intervals and the Constant-Returns Test

The three inferential questions, answered from the fitted model: **direction** (signs of $\hat\alpha,\hat\beta$), **magnitude** (the elasticities), and **confidence** (intervals and tests). The headline hypothesis: $H_0: \alpha + \beta = 1$ — constant returns to scale, exactly as the theory's baseline assumed.

```python
print('95% confidence intervals:')
print(model.conf_int().rename(columns={0: 'lower', 1: 'upper'}).round(3))

# Test H0: alpha + beta = 1 (constant returns to scale)
test = model.t_test('logK + logL = 1')
print(f'\nH0: alpha + beta = 1   vs   H1: alpha + beta != 1')
print(f't = {test.tvalue.item():.3f},  p-value = {test.pvalue.item():.3f}')
print('-> We FAIL to reject constant returns: the data are fully consistent')
print('   with the theory Cobb and Douglas wrote down before seeing them.')

# Impose the theory a priori: alpha + beta = 1  =>  per-worker form
restricted = smf.ols('np.log(Y / L) ~ np.log(K / L)', data=cd).fit()
alpha_r = restricted.params.iloc[1]
A_r = np.exp(restricted.params.iloc[0])
print(f'\nRestricted estimate (constant returns imposed):')
print(f'   Y = {A_r:.2f} * K^{alpha_r:.2f} * L^{1 - alpha_r:.2f}')
print('This is the celebrated  Y = 1.01 K^0.25 L^0.75  from the 1928 paper.')
```

Consequence: exponentiating results estimated in logs can <strong>bias</strong> conclusions back on the original scale. The log-linear form remains standard practice — interpretable and computationally convenient — but the retransformation issue is worth remembering.</p>

### Example 7 — Jensen's Inequality, Simulated

If $\log Y \sim \mathcal{N}(\mu, \sigma^2)$, then $\mathbb{E}[Y] = e^{\mu + \sigma^2/2}$ — strictly larger than $e^{\mu} = e^{\mathbb{E}[\log Y]}$. Naively exponentiating a log-scale mean **underestimates** the level; the correction factor $e^{\sigma^2/2}$ is known as the *smearing* adjustment.

```python
mu, sigma_j = 2.0, 0.8
logY_sim = rng.normal(mu, sigma_j, 200_000)
Y_sim = np.exp(logY_sim)

naive = np.exp(logY_sim.mean())          # exp(E[log Y])  <- Jensen trap
truth = Y_sim.mean()                     # E[Y]
corrected = np.exp(logY_sim.mean() + logY_sim.var() / 2)   # smearing fix

print(f'exp(E[log Y])  (naive)     : {naive:8.3f}')
print(f'E[Y]           (truth)     : {truth:8.3f}')
print(f'theory e^(mu + sigma^2/2)  : {np.exp(mu + sigma_j**2 / 2):8.3f}')
print(f'smearing-corrected         : {corrected:8.3f}')
print(f'\nThe naive back-transform misses E[Y] by '
      f'{100 * (truth / naive - 1):.1f}% — the price of ignoring Jensen.')
```

### Steps 1–3: Name, Translate, Classify

**Step 1 — Name the quantities.** Identify the **state variable**: the quantity that changes (call it $y$), and what it depends on (time $t$, or another variable $x$). Identify the **parameters**: fixed constants with real-world meaning (rates, capacities, elasticities). Write down **units** for everything — unit mismatches are the fastest way to catch a wrong equation.

**Step 2 — Translate assumptions into a rate.** Most modeling assumptions are statements about **how fast** something changes: "grows proportionally to …", "decays at rate …", "responds proportionally to …". Each phrase becomes a term in

$$\frac{dy}{dt}=\text{(sum of the mechanisms)}.$$

*Cobb–Douglas version:* "marginal product proportional to average product" became $\partial Y/\partial K=\alpha\,Y/K$.

**Step 3 — Classify the equation.**
- **Separable?** Can you write it as $g(y)\,dy=h(t)\,dt$? (Most models in this course are.)
- **Linear?** Form $y'+p(t)\,y=q(t)$ — solved with an integrating factor.

The classification decides the technique. This lecture: **separation of variables**, the workhorse.

### Steps 4–6: Separate, Pin Down, Interpret

**Step 4 — Separate and integrate.** Move everything with $y$ to the left, everything with $t$ to the right, and integrate:

$$\frac{dy}{g(y)}=h(t)\,dt
\quad\Longrightarrow\quad
\int\frac{dy}{g(y)}=\int h(t)\,dt+C.$$

**Don't forget $C$** — and if another variable was held fixed (as with $L$ in Cobb–Douglas), the "constant" may be a *function* of it.

**Step 5 — Apply the initial condition.** The general solution has a free constant; the **initial condition** $y(0)=y_0$ pins it down. Substitute $t=0$, solve for $C$, substitute back. Result: a **unique** function — the only one consistent with both the mechanism and the starting point.

**Step 6 — Interpret and validate.** Read each parameter's **meaning** from the solution (growth rate, saturation level, elasticity, …). **Sanity checks:** limits as $t\to 0$ and $t\to\infty$; signs; units; special parameter values. Only then bring in **data**: estimate the parameters, quantify uncertainty, test the theory's predictions.

### Mini-Example: Continuous Compounding

**Problem.** A savings balance $B(t)$ earns interest continuously at annual rate $r$. Model and solve.

- **Step 1 (quantities):** state $B(t)$ in dollars; parameter $r$ in $1/\text{year}$; initial deposit $B_0$.
- **Step 2 (rate):** "interest accrues in proportion to the current balance":
$$\frac{dB}{dt}=rB.$$
- **Step 3 (classify):** separable.
- **Step 4 (separate and integrate):**
$$\frac{dB}{B}=r\,dt\;\Longrightarrow\;\log B=rt+C.$$
- **Step 5 (initial condition):** $B(0)=B_0$ gives $C=\log B_0$, hence
$$B(t)=B_0e^{rt}.$$
- **Step 6 (interpret):** $r$ is the continuous growth rate; doubling time $=\log 2/r$; and $t\to 0$ recovers $B_0$. ✓

### Example 8 — Compounding: Analytic vs. Numerical Solution

We plot the closed-form solution $B(t) = B_0 e^{rt}$ and let `scipy`'s `solve_ivp` integrate $dB/dt = rB$ numerically from scratch. The two agree to machine precision — the calculus was right — and the doubling time lands where $\log 2 / r$ says it should.

```python
B0, r = 1000.0, 0.05                      # $1,000 at 5% per year
t_grid = np.linspace(0, 40, 400)
B_analytic = B0 * np.exp(r * t_grid)

sol = solve_ivp(lambda t, B: r * B, t_span=(0, 40), y0=[B0],
                t_eval=np.linspace(0, 40, 21), rtol=1e-10)

t_double = np.log(2) / r
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t_grid, B_analytic, color=BC_BLUE, linewidth=2.5,
        label='analytic  $B_0 e^{rt}$')
ax.plot(sol.t, sol.y[0], 'o', color=BC_RED, markersize=7,
        label='numerical  (solve_ivp)')
ax.axhline(2 * B0, color=BC_SILVER, linestyle=':')
ax.axvline(t_double, color=BC_SILVER, linestyle='--')
ax.annotate(f'doubling time $= \\log 2 / r = {t_double:.1f}$ years',
            xy=(t_double, 2 * B0), xytext=(t_double + 2, 1.35 * B0),
            arrowprops=dict(arrowstyle='->', color=BC_NAVY), color=BC_NAVY)
ax.set_xlabel('years')
ax.set_ylabel('balance ($)')
ax.set_title('Continuous Compounding: $dB/dt = rB \\Rightarrow B(t) = B_0 e^{rt}$')
ax.legend()
fig.text(0.99, -0.02, 'Source: CS 310 Lecture 3.', ha='right', fontsize=9, color=BC_SILVER)
plt.tight_layout()
plt.show()

max_gap = np.max(np.abs(sol.y[0] - B0 * np.exp(r * sol.t)))
print(f'Largest analytic-vs-numerical gap: {max_gap:.2e} dollars — the calculus checks out.')
```

- **Cobb–Douglas:** rate of change of $Y$ in $K$ proportional to $Y/K$ $\Rightarrow$ $A K^{\alpha}L^{\beta}$.

### Example 9 — Exponential Growth in Real Data: `AirPassengers.csv`

**Theory first:** if air travel grows at a constant relative rate, then $dP/dt = rP$, so $\log P$ is **linear in time** — the model is fixed before we look. We then estimate $r$ by ERM on **all 144 months** and read the residuals not as "test error" but as *departures from the theory* — which turn out to have a story of their own (seasonality).

```python
air = load_data('AirPassengers.csv')
air.columns = ['month', 'passengers']
air['t'] = np.arange(len(air)) / 12          # time in years since Jan 1949

growth = smf.ols('np.log(passengers) ~ t', data=air).fit()
r_hat = growth.params['t']
ci = growth.conf_int().loc['t']
print(f'Estimated continuous growth rate r = {r_hat:.4f} per year')
print(f'95% CI: [{ci[0]:.4f}, {ci[1]:.4f}]')
print(f'Doubling time = log(2)/r = {np.log(2) / r_hat:.2f} years')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(air.t + 1949, air.passengers, color=BC_NAVY, linewidth=1.5,
             label='monthly passengers')
axes[0].plot(air.t + 1949, np.exp(growth.fittedvalues), color=BC_RED,
             linewidth=2.5, label='$P_0 e^{\\hat{r}t}$ (theory)')
axes[0].set_xlabel('year')
axes[0].set_ylabel('passengers (thousands)')
axes[0].set_title('Air Travel 1949–1960: Exponential Growth')
axes[0].legend()

axes[1].plot(air.t + 1949, growth.resid, color=BC_BLUE, linewidth=1.5)
axes[1].axhline(0, color=BC_SILVER)
axes[1].set_xlabel('year')
axes[1].set_ylabel('residual (log scale)')
axes[1].set_title('Residuals: the Theory Misses a Seasonal Cycle')
fig.text(0.99, -0.02, 'Source: AirPassengers dataset, CS 310 repository.',
         ha='right', fontsize=9, color=BC_SILVER)
plt.tight_layout()
plt.show()

print('\nThe regular 12-month wave in the residuals is not noise — it is a')
print('mechanism the model omitted (summer travel). In the inferential paradigm,')
print('residual structure is a finding about the WORLD, and an invitation to')
print('refine the theory (e.g., add monthly seasonality).')
```

### Example 10 — Advertising Elasticity: A Theory with a Testable Prediction

**Theory first:** marketing theory holds that advertising has **diminishing returns** — each extra dollar of TV budget buys less additional sales than the last. A constant-elasticity law, $\text{sales} = A \cdot \text{TV}^{\beta}$, encodes this whenever $\beta < 1$ (same derivation as Cobb–Douglas, one input). So *before touching the data* the theory predicts $0 < \beta < 1$ — and that is a hypothesis we can test.

```python
adv = load_data('Advertising.csv', index_col=0)

elastic = smf.ols('np.log(sales) ~ np.log(TV)', data=adv).fit()
b_hat = elastic.params['np.log(TV)']
se_b = elastic.bse['np.log(TV)']
ci_b = elastic.conf_int().loc['np.log(TV)']

print(f'Estimated TV elasticity: beta_hat = {b_hat:.4f}')
print(f'95% CI: [{ci_b[0]:.4f}, {ci_b[1]:.4f}]')
print('-> a 1% increase in TV budget is associated with a '
      f'{b_hat:.2f}% increase in sales.')

# One-sided test of the theory's prediction:  H0: beta = 1  vs  H1: beta < 1
t_stat = (b_hat - 1) / se_b
p_one_sided = stats.t.cdf(t_stat, df=elastic.df_resid)
print(f'\nH0: beta = 1 (no diminishing returns)  vs  H1: beta < 1')
print(f't = {t_stat:.2f},  one-sided p-value = {p_one_sided:.2e}')
print('-> Diminishing returns CONFIRMED, exactly as the theory predicted.')

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(adv.TV, adv.sales, s=25, color=BC_SILVER, alpha=0.7, label='markets')
tv_grid = np.linspace(adv.TV.min(), adv.TV.max(), 200)
A_adv = np.exp(elastic.params['Intercept'])
ax.plot(tv_grid, A_adv * tv_grid**b_hat, color=BC_RED, linewidth=3,
        label=f'$\\widehat{{sales}} = {A_adv:.2f}\\,TV^{{{b_hat:.2f}}}$')
ax.set_xlabel('TV advertising budget ($1,000s)')
ax.set_ylabel('sales (1,000s of units)')
ax.set_title('Diminishing Returns to TV Advertising: The Curve Flattens')
ax.legend()
fig.text(0.99, -0.02, 'Source: Advertising dataset (ISLR), CS 310 repository.',
         ha='right', fontsize=9, color=BC_SILVER)
plt.tight_layout()
plt.show()
```

### Example 11 — Physics on the Road: Fuel Use and Weight

**Theory first:** the energy needed to move a car scales with its **mass**, so the fuel burned per mile should be (approximately) **linear in weight** — not `mpg`, which is a reciprocal and therefore *cannot* be linear if the physics is. The model $\text{GPM} = \beta_0 + \beta_1 \cdot \text{weight}$, with $\beta_1 > 0$, is fixed before we look at a single car.

```python
auto = load_data('Auto.csv', na_values='?').dropna(subset=['horsepower'])
auto['gpm'] = 100 / auto.mpg                       # gallons per 100 miles
auto['weight_klb'] = auto.weight / 1000            # weight in 1,000 lb

phys = smf.ols('gpm ~ weight_klb', data=auto).fit()
b1 = phys.params['weight_klb']
ci1 = phys.conf_int().loc['weight_klb']

print(phys.summary().tables[1])
print(f'\nEach extra 1,000 lb of car burns {b1:.2f} more gallons per 100 miles')
print(f'(95% CI: [{ci1[0]:.2f}, {ci1[1]:.2f}]) — positive, as the physics demanded.')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(auto.weight_klb, auto.gpm, s=18, color=BC_SILVER, alpha=0.7)
w_grid = np.linspace(auto.weight_klb.min(), auto.weight_klb.max(), 100)
axes[0].plot(w_grid, phys.params['Intercept'] + b1 * w_grid, color=BC_RED,
             linewidth=3)
axes[0].set_xlabel('weight (1,000 lb)')
axes[0].set_ylabel('gallons per 100 miles')
axes[0].set_title('Fuel Use Is Linear in Weight (theory scale)')

axes[1].scatter(auto.weight_klb, auto.mpg, s=18, color=BC_SILVER, alpha=0.7)
axes[1].plot(w_grid, 100 / (phys.params['Intercept'] + b1 * w_grid),
             color=BC_BLUE, linewidth=3)
axes[1].set_xlabel('weight (1,000 lb)')
axes[1].set_ylabel('miles per gallon')
axes[1].set_title('The SAME Model Is a Curve on the mpg Scale')
fig.text(0.99, -0.02, 'Source: Auto dataset (ISLR), CS 310 repository.',
         ha='right', fontsize=9, color=BC_SILVER)
plt.tight_layout()
plt.show()

print('\nNote the payoff of thinking first: one physics argument chose the')
print('response scale (GPM, not mpg) — no model search, no validation split.')
```

### Example 12 — The Life-Cycle Wage Profile: A Theory That Predicts a Sign

**Theory first:** human-capital theory says wages rise as workers accumulate skills, but at a **decreasing** rate, and eventually flatten or decline. That predicts a concave age profile — $\log(\text{wage}) = \beta_0 + \beta_1\,\text{age} + \beta_2\,\text{age}^2$ with $\beta_1 > 0$ and $\beta_2 < 0$ — **written down before the data**. (The quadratic is not a Lecture 2-style search over degrees: theory delivered it directly.)

```python
wage = load_data('Wage.csv')

lifecycle = smf.ols('np.log(wage) ~ age + I(age**2)', data=wage).fit()
b1, b2 = lifecycle.params['age'], lifecycle.params['I(age ** 2)']
p2 = lifecycle.pvalues['I(age ** 2)']

print(lifecycle.summary().tables[1])
print(f'\nPredicted signs: beta1 > 0 and beta2 < 0.')
print(f'Estimated:       beta1 = {b1:+.4f},  beta2 = {b2:+.6f}   '
      '-> both as predicted')
print(f'One-sided test of H0: beta2 = 0 vs H1: beta2 < 0: '
      f'p = {p2 / 2:.2e}  (concavity confirmed)')

peak_age = -b1 / (2 * b2)
print(f'Wages peak at age -beta1/(2 beta2) = {peak_age:.1f} years')

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(wage.age, wage.wage, s=8, color=BC_SILVER, alpha=0.4, label='workers')
age_grid = np.linspace(18, 80, 200)
wage_fit = np.exp(lifecycle.params['Intercept'] + b1 * age_grid + b2 * age_grid**2)
ax.plot(age_grid, wage_fit, color=BC_RED, linewidth=3, label='life-cycle profile')
ax.axvline(peak_age, color=BC_NAVY, linestyle='--', linewidth=1.5)
ax.annotate(f'peak at {peak_age:.0f}', xy=(peak_age, wage_fit.max()),
            xytext=(peak_age + 6, wage_fit.max() + 25), color=BC_NAVY,
            arrowprops=dict(arrowstyle='->', color=BC_NAVY))
ax.set_xlabel('age')
ax.set_ylabel('wage (thousands of USD)')
ax.set_title('The Concave Life-Cycle Wage Profile, Exactly as Theory Predicted')
ax.legend()
fig.text(0.99, -0.02, 'Source: Wage dataset (ISLR), CS 310 repository.',
         ha='right', fontsize=9, color=BC_SILVER)
plt.tight_layout()
plt.show()
```
