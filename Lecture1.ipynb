  <h2 style="color: #003D79; font-family: 'Georgia', 'Palatino Linotype', serif; margin-top: 0;">A Coin Toss Is (In Principle) Deterministic</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">With <strong>perfect knowledge</strong> of all physical factors — initial position, velocity, angular momentum, air resistance, surface interactions — we could predict a coin toss exactly. Under this idealization there exists an unknown function $g^{\star}(\mathbf{x})$ mapping the <em>complete</em> set of covariates $\mathbf{x} \in \mathbb{R}^p$ to the outcome (heads or tails). <strong>Covariates</strong> (<span style='color:#C4122F;'>independent / explanatory variables</span>) are all the physical factors that drive the outcome.</p>
</div>

<div style="border-left: 5px solid #003D79; background-color: #F5F7FA; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #003D79; font-family: 'Georgia', 'Palatino Linotype', serif; margin-top: 0;">From Deterministic to Probabilistic</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">In practice, it is <span style='color:#C4122F;'><strong>impossible</strong></span> to observe or measure <em>all</em> attributes of a real-world phenomenon. So instead of recovering the true deterministic $g^{\star}(\cdot)$, we adopt a <strong>probabilistic perspective</strong>: unobserved covariates are absorbed into a <strong>random component</strong>.</p>
</div>

<div style="background-color: #162F57; padding: 18px 22px; margin: 20px 0; border-radius: 8px;">
  <p style="margin: 0; color: #FFF; font-weight: 700; font-family: 'Georgia', 'Palatino Linotype', serif;">Key Idea</p>
  <p style="margin: 10px 0 0 0; color: #E8EEF5; line-height: 1.6;">Nature generates outcomes through an unknown function of the <strong>observable</strong> covariates, plus a <strong>random component</strong> capturing unobserved or intrinsically unpredictable influences.</p>
</div>

### Example — Two Students, Same Covariates

Two students study **4 hours each** and have the **same mathematical background**. Yet one scores $4/10$ and the other scores $8/10$. **Why?**

A deterministic model $g(x_1, x_2)$ where $x_1 =$ hours studied and $x_2 =$ math background predicts the **same** outcome for both. Reality deviates. The discrepancy arises from factors we do not observe:

- One student may have suffered a **migraine** before the exam.
- The other may have practiced problem types **closely aligned** with the exam content.
- **Fatigue**, **stress**, or pure **chance** may also play a role.

These influences are collectively absorbed into a random component $\boldsymbol{\varepsilon}$, which explains deviations between the predicted value and the realized outcome.

<div style="background-color: #FBEAED; border-left: 4px solid #C4122F; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #C4122F; font-weight: 700;">The Fundamental Assumption of Supervised Learning</p>
  <p style="margin: 10px 0 0 0; color: #333; text-align: center; font-size: 1.15em;">$$Y = g^{\star}(\mathbf{x}) + \boldsymbol{\varepsilon}$$</p>
  <p style="margin: 10px 0 0 0; color: #333;">$g^{\star}(\cdot)$: the <strong>systematic, learnable structure</strong> of the problem.<br>
  $\boldsymbol{\varepsilon}$: the <strong style="color:#C4122F;">irreducible uncertainty</strong> — unobserved factors, measurement limits, inherent variability.</p>
</div>

<div style="border-left: 5px solid #003D79; background-color: #F5F7FA; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #003D79; font-family: 'Georgia', 'Palatino Linotype', serif; margin-top: 0;">Definition: Data Generating Process (DGP)</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">A <strong>DGP</strong> is a probability model specifying how observations are produced. It consists of: a probability space $(\Omega, \mathcal{A}, \mathbb{P})$; an input set $X: \Omega \to \mathcal{X} \subseteq \mathbb{R}^p$; and an output random variable $Y: \Omega \to \mathbb{R}$ generated by $Y = g^{\star}(\mathbf{x}) + \varepsilon$, with $g^{\star}: \mathcal{X} \to \mathbb{R}$ unknown and measurable, and $\mathbb{E}[\varepsilon \mid X = \mathbf{x}] = 0$. The DGP describes the <strong>mechanism</strong> through which outcomes arise: partially observed structure + an irreducible random element.</p>
</div>

### A Concrete DGP: Tossing a Coin

**Phenomenon:** a coin is tossed and we record $Y \in \{0, 1\}$, where $1$ = heads and $0$ = tails.

- **Probability space** $(\Omega, \mathcal{A}, \mathbb{P})$: $\Omega$ = all possible toss scenarios — every combination of launch force, spin, air currents, landing surface, ...
- **Inputs** $X = (X_1, X_2)$: what our instruments *can* measure — $X_1$ = initial upward velocity, $X_2$ = spin rate.
- **Systematic part**: $g^{\star}(\mathbf{x}) = \mathbb{E}[Y \mid X = \mathbf{x}] = \mathbb{P}(\text{heads} \mid \mathbf{x})$. For a fair toss: $g^{\star}(\mathbf{x}) = 0.5$.
- **Noise** $\boldsymbol{\varepsilon}$: everything we did *not* measure — micro air currents, exact catch height, bounce on landing — with $\mathbb{E}[\varepsilon \mid X = \mathbf{x}] = 0$.
- **Output**: toss A lands heads: $Y = 0.5 + 0.5 = 1$; toss B lands tails: $Y = 0.5 - 0.5 = 0$.

**Same $\mathbf{x}$, different $Y$:** two tosses with identical measured velocity and spin can still land differently — only $\boldsymbol{\varepsilon}$ differs. With *complete* information, $\boldsymbol{\varepsilon}$ would vanish and the toss would be deterministic.

### Example 2 — Simulating the Coin-Toss DGP

```python
# Same measured covariates x, different outcomes Y: only epsilon differs
n_tosses = 20
g_star = 0.5                                  # P(heads | x) for a fair toss
Y = rng.binomial(1, g_star, size=n_tosses)    # outcomes
eps = Y - g_star                              # realized random component

df_toss = pd.DataFrame({'toss': range(1, n_tosses + 1),
                        'g_star(x)': g_star, 'epsilon': eps, 'Y': Y})
print(df_toss.head(10).to_string(index=False))
print(f"\nProportion of heads in {n_tosses} tosses: {Y.mean():.2f}  (g_star = {g_star})")
```

### Example 3 — Two Students, Simulated at Scale

We simulate many students who all share the *same* covariates (4 hours studied, same background). The deterministic part predicts the same score for everyone; the random component $\varepsilon$ spreads the outcomes.

```python
n_students = 200
hours = np.full(n_students, 4.0)          # identical covariate for everyone
g_star_score = 1.2 * hours + 1.5          # systematic part: g*(x) = 1.2x + 1.5 = 6.3
eps = rng.normal(0, 1.4, n_students)      # unobserved: migraines, luck, stress...
score = np.clip(g_star_score + eps, 0, 10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(score, bins=25, color=BC_BLUE, edgecolor='white', alpha=0.85)
ax.axvline(g_star_score[0], color=BC_RED, linewidth=2.5,
           label=f'Deterministic prediction $g^*(x)$ = {g_star_score[0]:.1f}')
ax.set_title('Same Covariates, Different Outcomes: the Role of $\\varepsilon$')
ax.set_xlabel('Exam score (0–10)')
ax.set_ylabel('Number of students')
ax.legend()
fig.text(0.99, 0.01, 'Source: simulated data, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
```

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">The universe — or, in practice, <strong>your boss</strong>, with almost comparable authority — hands you a <strong>tabular dataset</strong>. One column is the <strong>outcome</strong> $Y$ we wish to learn (supervised paradigm). The remaining columns are <strong>covariates</strong> $\mathbf{x} \in \mathbb{R}^p$ that may explain $Y$ through an unknown function $g^{\star}(\mathbf{x})$.</p>

### The Statistical Framework

- The joint distribution of $(\mathbf{X}, Y)$ induced by the DGP is **fixed but unknown**.
- The observed data $(\mathbf{X}_1, Y_1), \ldots, (\mathbf{X}_n, Y_n)$ are **i.i.d.** draws from this distribution.
- **Goal:** construct an estimator $\hat{g}$ from a **hypothesis class** $\mathcal{G}$, using finite data, so that $\hat{g} \approx g^{\star}$ with respect to a chosen **loss function**.
- The random component $\boldsymbol{\varepsilon}$ is <span style="color:#C4122F;">**irreducible**</span>: it sets a fundamental limit on achievable predictive accuracy.

### The Probability Space $(\mathcal{X}, \mathcal{A}, \mathbb{P})$

Let $X_1, X_2, \ldots$ be i.i.d. copies of a random variable $X$ on a probability space $(\mathcal{X}, \mathcal{A}, \mathbb{P})$, where:

- $\mathcal{X}$ — the **sample space**: all possible values the random variable $X$ can take.
- $\mathcal{A}$ — a **$\sigma$-algebra** on $\mathcal{X}$: which subsets of $\mathcal{X}$ are *measurable events* to which probabilities can be assigned.
- $\mathbb{P}$ — a **probability measure** defined on $(\mathcal{X}, \mathcal{A})$.

### The Borel $\sigma$-Algebra

When $(\mathcal{X}, \mathcal{A}) = (\mathbb{R}, \mathcal{B})$, the random variable takes values on the **real line**. $\mathcal{B}$ is the **smallest** $\sigma$-algebra containing all **open intervals** $(a, b)$ of $\mathbb{R}$ — and therefore also closed intervals, half-open intervals, and countable unions and intersections of intervals. This construction makes probabilities, expectations, and limiting operations involving real-valued random variables **mathematically well-defined**.

**Example — heights are Borel-measurable:** let $X$ = height (in meters) of a randomly selected individual. The event $E = \{\omega \in \Omega : 1.60 \leq X(\omega) \leq 1.90\}$ corresponds to the Borel set $[1.60, 1.90] \in \mathcal{B}$, so $\mathbb{P}(E)$ is **well-defined**.

### Loss Functions: Defining "Best"

A **loss function** is a measurable mapping $\ell: \mathcal{Y} \times \mathcal{Y} \to \mathbb{R}^+$, where $\ell(\hat{y}, y)$ is the **penalty** for predicting $\hat{y} = \hat{g}(\mathbf{x})$ when the true outcome is $y$. Common examples:

| Loss | Formula | Optimal constant predictor |
|:-----|:--------|:---------------------------|
| **Squared loss** | $\ell(\hat{y}, y) = (\hat{y} - y)^2$ | Sample **mean** |
| **Absolute loss** | $\ell(\hat{y}, y) = \lvert \hat{y} - y \rvert$ | Sample **median** |
| **Zero–one loss** | $\ell(\hat{y}, y) = \mathbf{1}\{\hat{y} \neq y\}$ | Sample **mode** |

- **Loss** $\ell(\hat{y}, y)$: *pointwise* — the penalty for a **single** prediction.
- **Risk** of a model $g$: *population-level* — the expected loss under the DGP, $R(g) = \mathbb{E}[\ell(g(\mathbf{X}), Y)]$.
- **Empirical risk** replaces the expectation with a finite-sample **average**: a data-driven proxy for the true risk.

### Empirical Risk Minimization for the Constant Model

With squared loss, the ERM problem is

$$\hat{\theta} = \underset{\theta \in \mathbb{R}}{\arg\min} \; \frac{1}{n} \sum_{i=1}^{n} (Y_i - \theta)^2.$$

Taking the derivative and setting it to zero gives the **closed-form solution**:

$$\hat{\theta} = \frac{1}{n} \sum_{i=1}^{n} Y_i = \bar{Y}, \qquad \hat{g}(x) = \bar{Y}.$$

### Example 4 — The Mean Minimizes the Empirical Squared Risk

We plot the empirical risk $\hat{R}_n(\theta)$ as a function of $\theta$ for the ten observed scores and verify the minimum sits exactly at $\bar{Y} = 7.346$.

```python
scores = np.array([8.32, 7.54, 5.65, 2.74, 10.00, 9.34, 6.76, 8.21, 7.56, 7.34])
theta_grid = np.linspace(0, 10, 500)
emp_risk = [(np.mean((scores - t) ** 2)) for t in theta_grid]

theta_hat = scores.mean()
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(theta_grid, emp_risk, color=BC_BLUE, linewidth=2.5,
        label='Empirical risk $\\hat{R}_n(\\theta)$')
ax.axvline(theta_hat, color=BC_RED, linestyle='--', linewidth=2,
           label=f'$\\bar{{Y}}$ = {theta_hat:.3f}')
ax.scatter([theta_hat], [np.mean((scores - theta_hat) ** 2)],
           color=BC_RED, zorder=5, s=80)
ax.set_title('Empirical Risk of the Constant Model (Squared Loss)')
ax.set_xlabel('Constant prediction $\\theta$')
ax.set_ylabel('Empirical risk')
ax.legend()
fig.text(0.99, 0.01, 'Source: 10 student scores, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
print(f'Best constant prediction for an unseen student: {theta_hat:.3f}')
```

### Example 5 — Changing the Loss Changes the Answer

Under **absolute** loss the optimal constant is the **median**, not the mean. Watch what happens when we add one extreme outlier.

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Under appropriate conditions, the sample mean $\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i$ converges <em>almost surely</em> to the population mean $\theta = \mathbb{E}[X]$: $\bar{X}_n \xrightarrow{\text{a.s.}} \theta$ as $n \to \infty$. <strong>In plain words:</strong> averaging enough data gets you (with probability one) to the truth.</p>

<div style="background-color: #FBEAED; border-left: 4px solid #C4122F; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #C4122F; font-weight: 700;">Theorem (Strong Law of Large Numbers)</p>
  <p style="margin: 10px 0 0 0; color: #333; text-align: center; font-size: 1.1em;">$$\text{If } \mathbb{E}[|X|] < \infty, \text{ then } \quad \frac{1}{n}\sum_{i=1}^{n} X_i \xrightarrow{\text{a.s.}} \mathbb{E}[X].$$</p>
</div>

**Why the Strong Law matters for learning:**

- In modern statistical learning, parameters are learned by minimizing **empirical objectives**: sample means of losses, gradients, or feature statistics from finite data.
- The Strong Law guarantees (under mild integrability) that these empirical quantities converge **almost surely** to their population counterparts as data grows.
- Consequences: minimizing **empirical risk** approximates minimizing **true risk**; **Monte Carlo** estimates converge to exact expectations; **stochastic optimization** behaves predictably in the large-sample limit.

The SLLN is the **mathematical bridge** between observed data and the theoretical distributions they are intended to represent — it is what *licenses* learning from data through empirical averages in the first place.

### Example 6 — Watching the SLLN in Action

Back to the ten students: increase the number of observed students $10 \to 50 \to 1{,}000 \to 100{,}000$ and watch the running sample mean approach the true population mean.

```python
true_theta = 7.0        # unknown population mean score (we play omniscient)
n_max = 100_000
draws = np.clip(rng.normal(true_theta, 1.8, n_max), 0, 10)
running_mean = np.cumsum(draws) / np.arange(1, n_max + 1)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(running_mean, color=BC_BLUE, linewidth=1.5, label='Running sample mean $\\bar{X}_n$')
ax.axhline(true_theta, color=BC_RED, linestyle='--', linewidth=2,
           label=f'Population mean $\\theta$ = {true_theta}')
for n in [10, 50, 1000]:
    ax.axvline(n, color=BC_SILVER, linestyle=':', linewidth=1)
    ax.annotate(f'n={n}', (n, ax.get_ylim()[0]), textcoords='offset points',
                xytext=(4, 10), fontsize=8, color=BC_NAVY)
ax.set_xscale('log')
ax.set_title('Strong Law of Large Numbers: $\\bar{X}_n \\to \\theta$ almost surely')
ax.set_xlabel('Sample size $n$ (log scale)')
ax.set_ylabel('Sample mean')
ax.legend()
fig.text(0.99, 0.01, 'Source: simulated scores, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
for n in [10, 50, 1000, 100_000]:
    print(f'n = {n:>7,}   sample mean = {running_mean[n-1]:.4f}   error = {abs(running_mean[n-1]-true_theta):.4f}')
```

### Modes of Convergence of Random Variables

To prove the SLLN — and to analyze learning algorithms in general — we must formalize how a sequence $\{X_n\}$ approaches a limiting random variable $X$. Three fundamental modes, from strongest to weakest:

1. **Almost sure convergence** — $\mathbb{P}\left(\left\{\omega \in \Omega : \lim_{n \to \infty} X_n(\omega) = X(\omega)\right\}\right) = 1$, denoted $X_n \xrightarrow{\text{a.s.}} X$. A *pathwise* notion: for almost every outcome $\omega$, the numerical sequence $\{X_n(\omega)\}$ converges to $X(\omega)$.
2. **Convergence in probability** — $\mathbb{P}(|X_n - X| > \varepsilon) \to 0$ for every $\varepsilon > 0$.
3. **Convergence in distribution** (in law) — the CDFs converge at continuity points; the mode behind the Central Limit Theorem.

They differ in strength, interpretation, and practical implications for asymptotic guarantees and inference.

### Stochastic Order Notation

In practice we rarely track exact limits alone — we also care about **rates**: how *quickly* random errors, estimation gaps, and stochastic fluctuations shrink as $n$ grows. The notation $O_p(\cdot)$ and $o_p(\cdot)$ expresses these rates probabilistically — the **stochastic analogues** of the deterministic Big-$O$ and little-$o$ notation. For the sample mean of i.i.d. data with finite variance: $\bar{X}_n - \theta = O_p(n^{-1/2})$.

### Example 7 — The Extension of the Strong Law: Transformed Sample Means

If $\mathbb{E}[|g(X)|] < \infty$, then $\frac{1}{n}\sum_{i=1}^n g(X_i) \xrightarrow{\text{a.s.}} \mathbb{E}[g(X)]$. This is the engine of **Monte Carlo**: we estimate $\mathbb{E}[X^2]$ for $X \sim \mathcal{N}(0,1)$ (true value $= 1$).

```python
n_max = 50_000
x = rng.normal(0, 1, n_max)
mc_estimate = np.cumsum(x ** 2) / np.arange(1, n_max + 1)   # g(X) = X^2

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(mc_estimate, color=BC_BLUE, linewidth=1.5,
        label='Monte Carlo estimate of $\\mathbb{E}[X^2]$')
ax.axhline(1.0, color=BC_RED, linestyle='--', linewidth=2,
           label='True value $\\mathbb{E}[X^2] = 1$')
# O_p(n^{-1/2}) envelope
n_axis = np.arange(1, n_max + 1)
ax.fill_between(n_axis, 1 - 2/np.sqrt(n_axis), 1 + 2/np.sqrt(n_axis),
                color=BC_SILVER, alpha=0.35, label='$O_p(n^{-1/2})$ envelope')
ax.set_xscale('log'); ax.set_ylim(0.5, 1.5)
ax.set_title('Transformed Sample Means Converge: $\\frac{1}{n}\\sum g(X_i) \\to \\mathbb{E}[g(X)]$')
ax.set_xlabel('Sample size $n$ (log scale)')
ax.set_ylabel('Estimate')
ax.legend()
fig.text(0.99, 0.01, 'Source: simulated N(0,1) draws, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
```

### Empirical Risk Converges to True Risk

The approximation $\hat{g}$ is typically obtained by minimizing the **empirical risk**

$$\hat{R}_n(g) = \frac{1}{n} \sum_{i=1}^{n} \ell\left(g(\mathbf{X}_i), Y_i\right).$$

**The fundamental justification:** under mild integrability, for any **fixed** model $g$,

$$\hat{R}_n(g) \xrightarrow{\text{a.s.}} R(g) = \mathbb{E}[\ell(g(X), Y)].$$

This pathwise convergence is the backbone of **PAC learning**: with enough data, empirical performance is a reliable proxy for generalization performance.

### Fixed Models vs. Data-Dependent Models

- A **fixed** model is *non-random*: chosen **before** observing the data (theory, domain expertise), then held fixed as $n$ grows — the classical statistical workflow.
- **Statistical learning reverses the order**: first observe the data, then select a hypothesis by minimizing empirical risk. The chosen model $\hat{g}_n$ now **depends on the sample**: it is a <span style="color:#C4122F;">**random object**</span>.
- Even in a parametric family $g_{\boldsymbol{\theta}}$: the *family* is fixed, but $\hat{\boldsymbol{\theta}}_n$ is computed from random data — so the learned predictor is a <span style="color:#C4122F;">**random function**</span>.

### Why the Strong Law Is Not Enough

The SLLN controls $\hat{R}_n(g) \to R(g)$ for *fixed* $g$ — it says **nothing** about $R(\hat{g}_n) - \hat{R}_n(\hat{g}_n)$, because $\hat{g}_n$ depends on the same sample. ERM examines **many** candidate hypotheses and picks the one that looks best on the observed sample. If $\mathcal{G}$ is rich enough, some model will fit sample-specific fluctuations unusually well: **very low empirical risk**, yet **large population risk**.

  <p style="margin: 0; color: #C4122F; font-weight: 700;">Overfitting</p>
  <p style="margin: 10px 0 0 0; color: #333;">The discrepancy between training performance and test performance caused by fitting <strong>noise</strong> instead of <strong>structure</strong>. <em>Underfitting</em> is the opposite extreme — a class too restrictive to capture genuine structure.</p>
</div>

### Example 8 — Overfitting in Action

We fit polynomials of increasing degree to a small sample from $Y = \sin(2\pi x) + \varepsilon$ and compare training vs. test error — very low empirical risk, large population risk.

```python
n_train, n_test = 15, 500
x_train = np.sort(rng.uniform(0, 1, n_train))
y_train = np.sin(2 * np.pi * x_train) + rng.normal(0, 0.25, n_train)
x_test = np.linspace(0, 1, n_test)
y_test = np.sin(2 * np.pi * x_test) + rng.normal(0, 0.25, n_test)

degrees = range(0, 13)
train_err, test_err = [], []
for d in degrees:
    coefs = np.polyfit(x_train, y_train, d)
    train_err.append(np.mean((np.polyval(coefs, x_train) - y_train) ** 2))
    test_err.append(np.mean((np.polyval(coefs, x_test) - y_test) ** 2))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(x_train, y_train, color=BC_NAVY, zorder=5, label='Training data')
xs = np.linspace(0, 1, 300)
for d, col in [(1, BC_SILVER), (3, BC_BLUE), (12, BC_RED)]:
    coefs = np.polyfit(x_train, y_train, d)
    axes[0].plot(xs, np.polyval(coefs, xs), color=col, linewidth=2, label=f'degree {d}')
axes[0].set_ylim(-2, 2)
axes[0].set_title('Fits of Increasing Complexity')
axes[0].set_xlabel('$x$'); axes[0].set_ylabel('$y$'); axes[0].legend()

axes[1].plot(list(degrees), train_err, 'o-', color=BC_BLUE, label='Training risk $\\hat{R}_n$')
axes[1].plot(list(degrees), test_err, 's-', color=BC_RED, label='Test risk $\\approx R$')
axes[1].set_yscale('log')
axes[1].set_title('Overfitting: Empirical Risk Lies')
axes[1].set_xlabel('Polynomial degree (model complexity)')
axes[1].set_ylabel('Mean squared error (log)')
axes[1].legend()
fig.text(0.99, 0.01, 'Source: simulated data, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.tight_layout(); plt.show()
```

### What We Actually Need: Uniform Convergence

ERM searches **globally** over the whole class — closeness at one value of $\theta$ is not enough. We need $\hat{R}_n \approx R$ **simultaneously for every** hypothesis:

$$\sup_{g \in \mathcal{G}} \left| \hat{R}_n(g) - R(g) \right| \xrightarrow{\mathbb{P}} 0.$$

Then the empirical risk surface is uniformly close to the true risk surface, so the empirical minimizer must lie near a true minimizer: no "spurious dips" can fool the optimizer — ERM cannot win merely by exploiting random fluctuations.

### Example 9 — Geometric Intuition: Empirical vs. Population Risk Curves

```python
theta = np.linspace(0.5, 6, 300)
pop_risk = 0.20 * (theta - 3.3) ** 2 + 1.0
emp_risk = 0.20 * (theta - 2.7) ** 2 + 1.05 + 0.12 * np.sin(1.7 * theta)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(theta, pop_risk, '--', color=BC_BLUE, linewidth=2.5,
        label='Population risk $R(\\theta)$')
ax.plot(theta, emp_risk, color=BC_RED, linewidth=2.5,
        label='Empirical risk $\\hat{R}_n(\\theta)$')
t_star = theta[np.argmin(pop_risk)]; t_hat = theta[np.argmin(emp_risk)]
ax.axvline(t_star, color=BC_BLUE, linestyle=':', linewidth=1.5)
ax.axvline(t_hat, color=BC_RED, linestyle=':', linewidth=1.5)
ax.annotate('$\\theta^*$', (t_star, 0.95), fontsize=14, color=BC_BLUE)
ax.annotate('$\\hat{\\theta}$', (t_hat, 0.95), fontsize=14, color=BC_RED)
ax.set_title('Uniform Closeness of Risk Curves Keeps $\\hat{\\theta}$ near $\\theta^*$')
ax.set_xlabel('$\\theta$'); ax.set_ylabel('Risk'); ax.legend()
fig.text(0.99, 0.01, 'Source: illustrative curves, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
```

### Step 1 — Concentration for a Fixed Model: Hoeffding's Inequality

Fix $g \in \mathcal{G}$ and define $Z_i(g) = \ell(g(X_i), Y_i)$, with bounded loss $0 \leq \ell \leq 1$. Then $\{Z_i(g)\}$ are i.i.d. and bounded with mean $R(g)$, and for any $\varepsilon > 0$:

$$\mathbb{P}\left(\left|\hat{R}_n(g) - R(g)\right| > \varepsilon\right) \leq 2\exp(-2n\varepsilon^2).$$

Empirical risk concentrates **exponentially fast** around the true risk. In rate notation: $\hat{R}_n(g) - R(g) = O_p(n^{-1/2})$ — the standard parametric rate.

### Step 2 — Finite Classes via the Union Bound

But $\hat{g}_n$ is not fixed — only $\mathcal{G}$ is. Suppose first $|\mathcal{G}| = N < \infty$. The union bound gives

$$\mathbb{P}\left(\sup_{g \in \mathcal{G}} \left|\hat{R}_n(g) - R(g)\right| > \varepsilon\right) \leq \sum_{g \in \mathcal{G}} \mathbb{P}\left(\left|\hat{R}_n(g) - R(g)\right| > \varepsilon\right) \leq 2N\exp(-2n\varepsilon^2).$$

The price of uniformity is only the factor $N$. Setting $2N\exp(-2n\varepsilon^2) \leq \delta$ and solving for $n$:

$$n \gtrsim \frac{1}{\varepsilon^2} \log\frac{N}{\delta} \;\; \Longrightarrow \;\; \text{with probability} \geq 1 - \delta, \; \sup_{g \in \mathcal{G}} \left|\hat{R}_n(g) - R(g)\right| \leq \varepsilon.$$

A **PAC-style guarantee** for finite classes: with enough samples, ERM generalizes reliably.

### Example 10 — Hoeffding's Bound vs. Reality

We check Hoeffding's inequality empirically for a fair-coin loss ($R(g) = 0.5$) and see how conservative the bound is.

```python
n_grid = np.arange(20, 2001, 20)
eps = 0.05
n_sims = 2000

emp_prob = []
for n in n_grid:
    samples = rng.binomial(1, 0.5, size=(n_sims, n)).mean(axis=1)
    emp_prob.append(np.mean(np.abs(samples - 0.5) > eps))
hoeffding = 2 * np.exp(-2 * n_grid * eps ** 2)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(n_grid, hoeffding, color=BC_RED, linewidth=2.5,
        label='Hoeffding bound $2e^{-2n\\varepsilon^2}$')
ax.plot(n_grid, emp_prob, color=BC_BLUE, linewidth=2,
        label='Empirical $\\mathbb{P}(|\\hat{R}_n - R| > \\varepsilon)$')
ax.set_ylim(0, 1.05)
ax.set_title(f'Concentration of the Empirical Risk ($\\varepsilon$ = {eps})')
ax.set_xlabel('Sample size $n$')
ax.set_ylabel('Probability of a large deviation')
ax.legend()
fig.text(0.99, 0.01, 'Source: 2,000 simulations per n, CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.show()
```

### Shattering and the VC Dimension

Let $\mathcal{G}$ be a class of $\{0,1\}$-valued functions on $\mathcal{X}$, and $A = \{x_1, \ldots, x_m\} \subset \mathcal{X}$ finite.

> $\mathcal{G}$ **shatters** $A$ if for *every* labeling $(b_1, \ldots, b_m) \in \{0,1\}^m$ there is some $g \in \mathcal{G}$ with $g(x_i) = b_i$ for all $i$.
>
> $\mathrm{VCdim}(\mathcal{G})$ = the **largest** $d$ such that *some* set of $d$ points is shattered ($= \infty$ if no maximum exists).

Useful fact: shattering $d$ points requires all $2^d$ labelings, so $\mathrm{VCdim}(\mathcal{G}) \leq \log_2 |\mathcal{G}|$ for finite classes.

**Three canonical examples:**

| Class | Definition | Free parameters | VC dimension |
|:------|:-----------|:----------------|:-------------|
| Thresholds on $\mathbb{R}$ | $g_t(x) = \mathbf{1}\{x \leq t\}$ | 1 | **1** |
| Intervals on $\mathbb{R}$ | $g_{a,b}(x) = \mathbf{1}\{x \in [a,b]\}$ | 2 | **2** |
| Axis-aligned rectangles in $\mathbb{R}^2$ | $g(x) = \mathbf{1}\{a \leq x_1 \leq b, \, c \leq x_2 \leq d\}$ | 4 | **4** |

- **Thresholds:** one point is shattered; two points fail — the labeling $(0, 1)$ with $x_1 < x_2$ is impossible, since labeling $x_1$ as $0$ forces $t < x_1$, which forces $x_2$ to be $0$ too.
- **Intervals:** two points are shattered; three fail — $(1, 0, 1)$ is impossible, since an interval containing $x_1$ and $x_3$ must also contain $x_2$ (the label-1 region has no gaps).
- **Rectangles:** four points in a "diamond" are shattered (each side controls exactly one point); five fail — one point lies inside the bounding box of the other four and gets trapped.

**A subtlety:** the VC dimension asks for the largest $d$ shattered in **some** configuration — not every configuration. (But beware: parameters = VC dimension is *not* a general rule.)

### Example 12 — Rectangles Shatter Four Points (the Diamond)

```python
from matplotlib.patches import Rectangle as MplRect
from itertools import product

pts = {'top': (2, 3.4), 'right': (3.4, 2), 'bottom': (2, 0.6), 'left': (0.6, 2)}
names = list(pts.keys())
coords = np.array(list(pts.values()))

fig, axes = plt.subplots(2, 4, figsize=(15, 7))
labelings = list(product([1, 0], repeat=4))[:8]   # show 8 of the 16 labelings
for ax, lab in zip(axes.ravel(), labelings):
    inside = coords[np.array(lab) == 1]
    if len(inside):
        (x0, y0), (x1_, y1_) = inside.min(axis=0) - 0.3, inside.max(axis=0) + 0.3
        ax.add_patch(MplRect((x0, y0), x1_ - x0, y1_ - y0,
                             facecolor=BC_BLUE, alpha=0.15,
                             edgecolor=BC_BLUE, linewidth=2))
    for (px, py), l in zip(coords, lab):
        ax.scatter(px, py, s=110, zorder=5,
                   facecolor=BC_NAVY if l else 'white',
                   edgecolor=BC_NAVY, linewidth=2)
    ax.set_xlim(-0.5, 4.5); ax.set_ylim(-0.5, 4.5)
    ax.set_xticks([]); ax.set_yticks([])
    ax.set_title(str(lab), fontsize=10)
fig.suptitle('Axis-Aligned Rectangles Realize Every Labeling of the Diamond'
             ' — VCdim = 4 (filled = label 1)', color=BC_BLUE, fontweight='bold')
fig.text(0.99, 0.01, 'Showing 8 of the 16 labelings. Source: CS 310 Lecture 1',
         ha='right', fontsize=8, color=BC_SILVER)
plt.tight_layout(); plt.show()
```

### Uniform Convergence for VC Classes

If $\mathrm{VC}(\mathcal{G}) = d < \infty$ and the loss is bounded, there exist universal constants $C, c > 0$ such that

$$\mathbb{P}\left(\sup_{g \in \mathcal{G}} \left|\hat{R}_n(g) - R(g)\right| > \varepsilon\right) \leq C\, n^d \exp(-c\, n \varepsilon^2).$$

Compare with the finite case: the factor $2N$ has been replaced by the polynomial $C n^d$. Despite infinitely many hypotheses, large deviations decay **exponentially** in $n$, up to a polynomial factor controlled by $d$ — and for fixed $\varepsilon$, the exponential eventually crushes $n^d$. Inverting the bound:

$$\sup_{g \in \mathcal{G}} \left|\hat{R}_n(g) - R(g)\right| = O_p\left(\sqrt{d/n}\right) = o_p(1).$$

The VC dimension plays the role that $\log N$ played for finite classes: it measures the **effective capacity** of $\mathcal{G}$. Finite VC dimension makes an infinite class behave like a finite one whose *effective size* grows only **polynomially** in $n$ — exactly what uniform convergence needs.

### From Uniform Convergence to a PAC Bound

Suppose the uniform convergence event holds: $\sup_{g} |\hat{R}_n(g) - R(g)| \leq \varepsilon$ with high probability. The argument is a chain of three inequalities — uniform convergence (UC), the definition of ERM, then UC again:

$$R(\hat{g}_n) \overset{\text{UC}}{\leq} \hat{R}_n(\hat{g}_n) + \varepsilon \overset{\text{ERM}}{\leq} \inf_{g \in \mathcal{G}} \hat{R}_n(g) + \varepsilon \overset{\text{UC}}{\leq} \inf_{g \in \mathcal{G}} R(g) + 2\varepsilon.$$

Each use of UC costs one $\varepsilon$ — using it twice is why the final bound has $2\varepsilon$.

<div style="background-color: #FBEAED; border-left: 4px solid #C4122F; padding: 15px 20px; margin: 20px 0; border-radius: 5px;">
  <p style="margin: 0; color: #C4122F; font-weight: 700;">The Canonical PAC Bound</p>
  <p style="margin: 10px 0 0 0; color: #333;">With probability at least $1 - \delta$ (for $n$ large enough relative to the complexity of $\mathcal{G}$),</p>
  <p style="margin: 10px 0 0 0; color: #333; text-align: center; font-size: 1.1em;">$$R(\hat{g}_n) - \inf_{g \in \mathcal{G}} R(g) \leq 2\varepsilon.$$</p>
  <p style="margin: 10px 0 0 0; color: #333;"><strong>P</strong>robably ($1-\delta$) <strong>A</strong>pproximately ($2\varepsilon$) <strong>C</strong>orrect: we cannot guarantee $\hat{g}_n$ is exactly optimal, nor that it works on every sample — but with high confidence it is close to the best the class can do.</p>
</div>
